```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method='spearman') # spearman is better for categorical data

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))

heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", square=True, cbar_kws={"label": "Spearman correlation"})

heatmap.set_title("Correlation Heatmap")
plt.show()



""" Observations:
- There is no strong correlation between the features. So no visible relationship. So delete observations with missing values.""" 
```
